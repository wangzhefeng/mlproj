```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from utils import gen_time_fea
```

# 1.入库流量数据

```python
ruku_df = pd.read_excel("data/入库流量数据.xlsx")
ruku_df.index = pd.to_datetime(ruku_df["TimeStample"])
del ruku_df["TimeStample"]
ruku_df = ruku_df.resample("3h").mean().interpolate('linear')
ruku_df.head(16)
```

# 2.遥测数据提取

```python
yaoce_df = pd.read_excel("data/遥测站降雨数据.xlsx")
yaoce_df.index = pd.to_datetime(yaoce_df["TimeStample"])
del yaoce_df["TimeStample"]
yaoce_df = yaoce_df.resample("3h").mean()
yaoce_df.head(16)
```

# 3.环境表

```python
huanjing_df = pd.read_excel("data/环境表.xlsx")
huanjing_df.index = pd.to_datetime(huanjing_df["TimeStample"])
del huanjing_df["TimeStample"]
huanjing_df = huanjing_df.resample("3h").mean().interpolate('bfill', limit_direction = "backward").fillna(method="ffill")
huanjing_df.head(16)
```

# 4.降雨预报数据

```python
jiangyu_df = pd.read_excel("data/降雨预报数据.xlsx")
jiangyu_df.index = pd.to_datetime(jiangyu_df["TimeStample"])
del jiangyu_df["TimeStample"]
jiangyu_df = jiangyu_df.resample("3h").mean().interpolate('bfill', limit_direction = "backward").fillna(method="ffill")
jiangyu_df.head(16)
```

# 5.数据合并

```python
ruku_df1 = gen_time_fea(ruku_df[:"2018-01-31"])
yaoce_df1 = yaoce_df[:"2018-01-31"]
huanjing_df1 = huanjing_df[:"2018-01-31"]
jiangyu_df1 = jiangyu_df[:"2018-01-31"]
data1 = pd.concat([ruku_df1, yaoce_df1, huanjing_df1, jiangyu_df1], axis = 1).dropna(axis = 0)
data1.tail()
```

```python
ruku_df2 = gen_time_fea(ruku_df["2018-07-01":"2018-07-31"])
yaoce_df2 = yaoce_df["2018-07-01":"2018-07-31"]
huanjing_df2 = huanjing_df["2018-07-01":"2018-07-31"]
jiangyu_df2 = jiangyu_df["2018-07-01":"2018-07-31"]

data2 = pd.concat([ruku_df2, yaoce_df2, huanjing_df2, jiangyu_df2], axis = 1).dropna(axis = 0)
data2.tail()
```

```python
ruku_df3 = gen_time_fea(ruku_df["2018-10-01":"2018-10-31"])
yaoce_df3 = yaoce_df["2018-10-01":"2018-10-31"]
huanjing_df3 = huanjing_df["2018-10-01":"2018-10-31"]
jiangyu_df3 = jiangyu_df["2018-10-01":"2018-10-31"]

data3 = pd.concat((ruku_df3, yaoce_df3, huanjing_df3, jiangyu_df3), axis=1).dropna(axis=0)
data3.tail()
```

## 测试数据

```python
data1["label"] = data1["Qi"].shift(-56)
data2["label"] = data2["Qi"].shift(-56)
data3["label"] = data3["Qi"].shift(-56)

data1_test = data1[-56:]
del data1_test["label"]
data2_test = data2[-56:]
del data2_test["label"]
data3_test = data3[-56:]
del data3_test["label"]
```

## 训练数据

```python
data1_train = data1[: -56]
data2_train = data2[: -56]
data3_train = data3[: -56]
data_train = pd.concat((data1_train, data2_train, data3_train), axis=0)
data_train.head()
```

## 数据文件保存

```python
data_train.to_csv("all_train_data.csv")
data1_test.to_csv("test_data_month1.csv")
data2_test.to_csv("test_data_month7.csv")
data3_test.to_csv("test_data_month10.csv")
```
