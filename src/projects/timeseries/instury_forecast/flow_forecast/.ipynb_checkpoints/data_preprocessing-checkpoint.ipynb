```python
import time
import datetime
import scipy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
from utils import gen_time_features
```

# 1 数据读取

## 1.1 入库流量数据

```python
ruku_df_chu = pd.read_excel("data/初赛/入库流量数据.xlsx")
ruku_df_jue = pd.read_excel("data/决赛/入库流量数据.xlsx")
ruku_df = pd.concat([ruku_df_chu, ruku_df_jue], axis = 0)
ruku_df.index = pd.to_datetime(ruku_df["TimeStample"])
del ruku_df["TimeStample"]
ruku_df = ruku_df.resample("3h").mean().interpolate('linear')
print(ruku_df.head(16))
print(ruku_df.shape)
```

## 1.2 遥测数据提取

```python
yaoce_df_chu = pd.read_excel("data/初赛/遥测站降雨数据.xlsx")
yaoce_df_jue = pd.read_excel("data/决赛/遥测站降雨数据.xlsx")
yaoce_df = pd.concat([yaoce_df_chu, yaoce_df_jue], axis = 0)
yaoce_df.index = pd.to_datetime(yaoce_df["TimeStample"])
del yaoce_df["TimeStample"]
yaoce_df = yaoce_df.resample("3h").mean()
# yaoce_df = yaoce_df.resample("3h").sum()
```

```python
# 求得当天遥测站降雨量和、均值
yaoce_df["all_rain"] = yaoce_df[["R" + str(i + 1) for i in range(yaoce_df.shape[1] - 1)]].sum(axis = 1)
yaoce_df["mean_rain"] = yaoce_df[["R" + str(i + 1) for i in range(yaoce_df.shape[1] - 1)]].mean(axis = 1)
print(yaoce_df.head())
print(yaoce_df.shape)
```

## 1.3 环境表

```python
huanjing_chu_df = pd.read_excel("data/初赛/环境表.xlsx")
huanjing_jue_df = pd.read_excel("data/决赛/环境表.xlsx")
huanjing_df = pd.concat([huanjing_chu_df, huanjing_jue_df], axis = 0)
huanjing_df.index = pd.to_datetime(huanjing_df["TimeStample"])
del huanjing_df["TimeStample"]
huanjing_df = huanjing_df.resample("3h").mean().interpolate('bfill', limit_direction = "backward").fillna(method = "ffill")
print(huanjing_df.head())
print(huanjing_df.shape)
```

## 1.4 降雨预报数据

```python
jiangyu_chu_df = pd.read_excel("data/初赛/降雨预报数据.xlsx")
jiangyu_jue_df = pd.read_excel("data/决赛/降雨预报数据.xlsx")
jiangyu_df = pd.concat([jiangyu_chu_df, jiangyu_jue_df], axis = 0)
jiangyu_df.index = pd.to_datetime(jiangyu_df["TimeStample"])
del jiangyu_df["TimeStample"]
jiangyu_df = jiangyu_df \
    .resample("3h") \
    .mean() \
    .interpolate('bfill', limit_direction = "backward") \
    .fillna(method = "ffill")
print(jiangyu_df.head())
print(jiangyu_df.shape)
```

## 1.5 数据合并

```python
ruku_df1 = gen_time_features(ruku_df[:"2018-01-31"])
yaoce_df1 = yaoce_df[:"2018-01-31"]
huanjing_df1 = huanjing_df[:"2018-01-31"]
jiangyu_df1 = jiangyu_df[:"2018-01-31"]
data1 = pd.concat([ruku_df1, yaoce_df1, huanjing_df1, jiangyu_df1], axis = 1).dropna(axis = 0)
print(data1.head()) 
print(data1.shape)
```

```python
ruku_df2 = gen_time_features(ruku_df["2018-07-01":"2018-07-31"])
yaoce_df2 = yaoce_df["2018-07-01":"2018-07-31"]
huanjing_df2 = huanjing_df["2018-07-01":"2018-07-31"]
jiangyu_df2 = jiangyu_df["2018-07-01":"2018-07-31"]

data2 = pd.concat([ruku_df2, yaoce_df2, huanjing_df2, jiangyu_df2], axis = 1).dropna(axis = 0)
print(data2.head())
print(data2.shape)
```

```python
ruku_df3 = gen_time_features(ruku_df["2018-10-01":"2018-10-31"])
yaoce_df3 = yaoce_df["2018-10-01":"2018-10-31"]
huanjing_df3 = huanjing_df["2018-10-01":"2018-10-31"]
jiangyu_df3 = jiangyu_df["2018-10-01":"2018-10-31"]

data3 = pd.concat((ruku_df3, yaoce_df3, huanjing_df3, jiangyu_df3), axis = 1).dropna(axis = 0)
print(data3.head())
print(data3.shape)
```
